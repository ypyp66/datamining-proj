```python
import numpy as np
import pandas as pd
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, SimpleRNN, Activation
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

hw = pd.read_csv('D:\datamining\datamining-proj-master\전년도_65세이상인구비율.csv', encoding='CP949')
hw = hw.fillna(0)
X = hw.iloc[1:, 2:-1] #2011~2019년 데이터
Y = hw.iloc[1:, -1:] #2020년 데이터

X = np.array(X)
X = X.reshape(278, 9, -1)
Y = np.array(Y)

model = Sequential()
model.add(SimpleRNN(50, return_sequences=False, input_shape=(4, 1)))
model.add(Dense(1))
model.summary()
model.compile(loss='mse',
              optimizer='adam',
              metrics=['accuracy'])
model.fit(X, Y, epochs=200, verbose=2)
print(model.predict(X))
Y_ = model.predict(X)
print(Y_.shape)
print(Y.shape)

line_fitter = LinearRegression()
x = np.linspace(0, 40, 100)
y = x
plt.plot(Y, Y, '.')
plt.plot(Y, Y_,'.')
plt.show()
```
