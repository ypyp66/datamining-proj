```python
location = eval(input("지역 입력(ex. 서울, 경기도, 부산, 전라남도, 경상북도, 충청도)"))
year = input("년도 입력(ex. 2012): ")

new_df=location[['행정구역',year+'년_계_총인구수',year+'년_계_연령구간인구수']]
new_df['65세이상 인구 비율']=location[year+'년_계_연령구간인구수']/location[year+'년_계_총인구수'] * 100
new_df
```

```python
location_ = input("지역구 입력: ")
location_df = location[location['행정구역'].str.contains(location_)]
for i in range(2011,2021):
    location_df[str(i)+'년65세이상비율'] = location_df.apply(lambda row: (row[str(i)+'년_계_연령구간인구수'] / row[str(i)+'년_계_총인구수']),axis=1)
over65 = location_df.iloc[:,-10:].T
over65.columns = ["비율"]
over65.index.name="년도"
over65.pivot_table(index = "년도", values="비율")
over65.plot(kind='barh', figsize=(10,10))
```

```python
import folium
font_name=font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
rc('font',family=font_name)
geo_data = 'TL_SCCO_SIG.geojson'
center = [37.541, 126.986]
# 맵이 center 에 위치하고, zoom 레벨은 11로 시작하는 맵 m을 만듭니다.
m = folium.Map(location=center, zoom_start=10)
# Choropleth 레이어를 만들고, 맵 m에 추가합니다.
folium.Choropleth( geo_data=geo_data, data = new_df, columns=('행정구역', '65세이상 인구 비율'),
 key_on='feature.properties.SIG_KOR_NM',
 fill_color='BuPu',
 legend_name='노령 인구비율' ).add_to(m)
m
```
